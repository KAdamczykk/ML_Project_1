```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

Podział zbioru danych dla modelarzy i walidatorów

```python
df = pd.read_csv("../data/NY-House-Dataset.csv")

learning_data, validation_data = train_test_split(df, test_size=0.3, random_state=42)

learning_data.to_csv("C../data/learning_data.csv", index=False)
validation_data.to_csv("../validation_data/validation_data.csv", index=False)
```
